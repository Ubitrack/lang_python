Introduction to UbiTrack Python Bindings
========================================

Let's first import some system packages

```python
import numpy as np
import time
import os
```

The Ubitrack bindings currently consists of the following modules:

- ubitrack.core.utmath
- ubitrack.core.utmeasurement
- ubitrack.dataflow.dataflow
- ubitrack.facade.facade

others might follow, to allow the use of algorithms and components directly.

So .. import some basic functionality:

```python
from ubitrack.core import math as utmath
from ubitrack.core import util
from ubitrack.facade import facade
```

In order to see what's happening, we configure logging:

```python
util.initLogging("tests/log4cpp.conf")
```

then we create an instance of AdvancedFacade:

```python
f = facade.AdvancedFacade("/usr/local/lib/ubitrack")
```

and load a simple dataflow, which produces test poses:

```python
f.loadDataflow("tests/single_pushsinkpose.dfg", True)
```

then we define a simple callback to receive measurements. Callbacks have a single argument - the measurement, which consists of a timestamp and a payload

```python
def cb(m):
    print m
```

then register the callback with the application push sink:

```python
f.setCallbackPose("receiver", cb)
```

next, we'll start the dataflow

```python
f.startDataflow()
```

now, the dataflow is running.

```python
time.sleep(3)
```

```python
f.stopDataflow()
```
